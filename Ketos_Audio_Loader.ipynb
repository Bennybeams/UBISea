Configuration of the audio representation and creation of the Spectrogram with **AudioLoader**. Parameters will be use to display the spectrogram of the sound.

Available parameters are :
- frame(float) : size of each segment of the sample
- step(float) : separation between segments. If None, then step=frame
- path(str) : specify path to the folder where wav files are stored if different than home folder
- filename (str) : specify wav files to be used. If not set all files from the folder will be used.
- channel (int) : on stereo recording, this option allows to choose left or right channel
- annotations (pandas DataFrame) : DF to be used for annotations on the sound if needed
- batch-size (int or str) : can be used to buffer the sound to reduce computational time. Default is 1, however it's possible to specify the size of a file to load it entirely before analyse.

num() get the number of segment of the spectrogram.

```python
rep = {'type':'MagSpectrogram', 
       'window':0.2, 
       'step':0.02, 
       'window_func':'hamming', 
       'freq_max':1000.}

loader = AudioFrameLoader(frame=30., step=15., filename=filename, repres=rep)
print(loader.num())
```

Plotting of the 1st segment of the file

```python
spec = next(loader)
spec.plot()
```
